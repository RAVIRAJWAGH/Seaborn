```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a bar plot with Seaborn
sns.set(style="whitegrid")
tips = sns.load_dataset("tips")
ax = sns.barplot(x="day", y="total_bill", data=tips, ci=None, palette="Set2") #red,orange, brown, yellow, purple, violet, pink

# Customization options
ax.set(title="Customized Bar Plot", xlabel="Day of the Week", ylabel="Total Bill")
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)  # Rotate x-axis labels
plt.ylim(0, 60)  # Adjust y-axis limits
plt.show()
```

# histogram

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a histogram using Seaborn
sns.set(style="whitegrid")
tips = sns.load_dataset("tips")
sns.histplot(data=tips, x="total_bill", kde=False, bins=20)  # kde=False disables the kernel density estimation
plt.title("Histogram of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

# kernel density plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a kernel density plot using Seaborn
sns.set(style="whitegrid")
tips = sns.load_dataset("tips")
sns.kdeplot(data=tips, x="total_bill", shade=True)  # shade=True shades the area under the curve
plt.title("Kernel Density Plot of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Density")
plt.show()
```

# Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load a dataset (e.g., tips dataset)
tips = sns.load_dataset("tips")

# Calculate the correlation matrix
correlation_matrix = tips.corr()

# Create a correlation heatmap
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
l=["viridis", "plasma", "inferno", "magma", "cividis", "coolwarm", "YlGnBu", "RdBu", "YlOrRd", "BuPu", "Greens",  "Oranges", "Reds", "cool", "hot", "PiYG", "PRGn", "BrBG", "seismic", "RdGy"]
import seaborn as sns
import matplotlib.pyplot as plt

# Load a dataset (e.g., tips dataset)
tips = sns.load_dataset("tips")

# Calculate the correlation matrix
correlation_matrix = tips.corr()
r=[i for i in range(1,21)]
for i,j in zip(l,r):
    # Create a correlation heatmap
    #plt.subplot(5,4,j)
    sns.heatmap(correlation_matrix, annot=True, cmap=i)
    plt.title(f"Correlation Heatmap with {i}")
    plt.show()
```
