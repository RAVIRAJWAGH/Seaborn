# Seaborn - "tips"dataset

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
tips=sns.load_dataset("tips")
tips
```

```python
sns.set(style="darkgrid")
sns.relplot(x="total_bill",y="tip",data=tips,hue="sex",col="day")
```

#### Q1) Find the females those are not smocking and they are giving tips 1:3 and show the seaborn digram.have total bill and time 

```python
q1=tips.loc[(tips["sex"]=="Female")&(tips["smoker"]=="No")&((tips["tip"]>1)&(tips["tip"]<3))]
q1
```

```python
sns.relplot(x="tip",y="total_bill",data=q1,hue="time")
```

#### Q2) Find the male and female those are having total_bill 10:20. They should visited on sat and sun .show seaborn graph with total bill and tips on smokers

```python
q2=tips.loc[((tips["total_bill"]>=10)&(tips["total_bill"]<=20))&((tips["day"]=="Sat")|(tips["day"]=="Sun"))]
q2
```

```python
q2["sex"].unique()
```

```python
q2["day"].unique()
```

```python
sns.relplot(x="total_bill",y="tip",data=q2,hue="day")
```

```python
sns.relplot(x="total_bill",y="tip",data=tips,hue="size",style="sex",col="smoker")
```

#### Q3) Find the male who gives tip in between 1:3 at dinner. And show seaborn graph of total bill and tips by size , day and smoking type.

```python
c=tips.loc[(tips["sex"]=="Female")&((tips["tip"]>=1)&(tips["tip"]<=3))&(tips["time"]=="Dinner")]
c
```

```python
sns.relplot(x="total_bill",y="tip",data=c,hue="day",size="smoker",style="day")
```
